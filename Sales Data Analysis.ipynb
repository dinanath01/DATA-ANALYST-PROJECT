```python
#importing necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('sales_data_sample_new.txt') #reading file
```

```python
df.head()#to see top 5 rows
```

```python
df.tail()#to see last 5 rows
```

```python
# Checking for missing values
df.isnull().sum()
```

```python
df.dtypes #checking datatypes
```

```python
#dealing with null values
df['ADDRESSLINE2']=df['ADDRESSLINE2'].fillna(df['ADDRESSLINE2'].mode()[0])
df['STATE']=df['STATE'].fillna(df['STATE'].mode()[0])
df['TERRITORY']=df['TERRITORY'].fillna(df['TERRITORY'].mode()[0])
df['POSTALCODE']=df['POSTALCODE'].fillna(df['POSTALCODE'].mode()[0])
```

```python
df.isnull().sum() #checking null values
```

```python
#Looking for null values if any, in heatmap
sns.heatmap(df.isna())
plt.show()
```

```python
df.drop_duplicates(inplace = True)
```

```python
df.describe()
```

```python
df['TOTALSALES'] = df['PRICEEACH'] * df['QUANTITYORDERED'] #creating a new column
```

```python
df['TOTALSALES']
```

```python
df.drop('Total Sales', axis=1, inplace=True)
```

```python
df
```

```python
# Create a 'MONTH_YEAR' column
df['ORDERDATE'] = pd.to_datetime(df['ORDERDATE'])
df['MONTH_YEAR'] = df['ORDERDATE'].dt.to_period('M')
```

```python
df
```

```python
# Top 5 Products by Sales
top_products = df.groupby("PRODUCTLINE")["SALES"].sum().sort_values(ascending=False)
plt.figure(figsize=(10,5))
sns.barplot(x=top_products.index, y=top_products.values)
plt.title("Total Sales by Product Line")
plt.ylabel("Sales")
plt.xticks(rotation=45)
plt.show()
```

```python
# Sales Trend Over Time
monthly_sales = df.groupby('MONTH_YEAR')["SALES"].sum()
monthly_sales.plot(kind='line', marker='o', figsize=(12, 6), title="Monthly Sales Trend")
plt.ylabel("Sales")
plt.grid(True)
plt.xticks(rotation=45)
plt.show()
```

```python
# Region-wise Sales
region_sales = df.groupby("COUNTRY")["SALES"].sum().sort_values(ascending=False).head(10)
plt.figure(figsize=(10,5))
sns.barplot(x=region_sales.index, y=region_sales.values)
plt.title("Top 10 Countries by Sales")
plt.ylabel("Sales")
plt.xticks(rotation=45)
plt.show()
```

```python
summary = df.groupby(["PRODUCTLINE", "COUNTRY"])["SALES"].sum().reset_index()
summary.to_excel("sales_summary_output.xlsx", index=False)
```
